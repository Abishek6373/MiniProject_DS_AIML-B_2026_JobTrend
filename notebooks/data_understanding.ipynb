# Stage 1: Data Understanding

## Project: JobTrend - Data Analysis of Employment Market Trends

### Objective
In this stage, we explore the structure, content, and quality of the LinkedIn job postings dataset to understand how it can be used to identify trending skills.

### Dataset Overview
The dataset consists of three primary files:
1. `linkedin_job_postings.csv`: Metadata about the job (Title, Company, Location, Level, Type).
2. `job_skills.csv`: Extracted skills associated with each job link.
3. `job_summary.csv`: Full job descriptions (Large file, ~4.8GB).

### Data Inspection

```python
import pandas as pd
import os

# Paths
raw_data_path = '../dataset/raw_data/'
metadata_file = os.path.join(raw_data_path, 'linkedin_job_postings.csv')
skills_file = os.path.join(raw_data_path, 'job_skills.csv')

# File sizes
for f in os.listdir(raw_data_path):
    size_gb = os.path.getsize(os.path.join(raw_data_path, f)) / (1024**3)
    print(f"{f}: {size_gb:.2f} GB")
```

### Loading Samples
Given the large size of the data, we start by analyzing a sample of 10,000 rows.

```python
df_meta = pd.read_csv(metadata_file, nrows=10000)
df_skills = pd.read_csv(skills_file, nrows=10000)

print("Metadata Columns:", df_meta.columns.tolist())
print("Skills Columns:", df_skills.columns.tolist())
```

### Preliminary Distributions
We look at the job levels and types available in the sample.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.countplot(data=df_meta, y='job_level', order=df_meta['job_level'].value_counts().index)
plt.title('Distribution of Job Levels')
plt.show()
```

### Initial Observations
- The dataset is heavily skewed towards 'Mid senior' levels in the initial sample.
- Most jobs listed are 'Onsite'.
- `job_link` serves as the primary key to join metadata with skills.
