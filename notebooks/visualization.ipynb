# Stage 4 & 5: Exploratory Data Analysis and Visualization

## Project: JobTrend

### Objective
Analyze the cleaned LinkedIn job data to identify key trends in the employment market, focusing on skill demand, job levels, and geographic distribution.

### Key Questions
1. Which skills are currently in highest demand?
2. How are job opportunities distributed across different seniority levels?
3. Which locations are the primary hubs for tech hiring?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load cleaned data
df = pd.read_csv('../dataset/processed_data/cleaned_job_data.csv')
print(f"Loaded {len(df)} records for analysis.")
```

### 1. Trending Skills Analysis
We extract and count the occurrences of individual skills.

```python
from collections import Counter

all_skills = []
for skills in df['cleaned_skills'].dropna():
    all_skills.extend([s.strip() for s in skills.split(',')])

skill_counts = Counter(all_skills).most_common(15)
skill_df = pd.DataFrame(skill_counts, columns=['Skill', 'Count'])

plt.figure(figsize=(12, 6))
sns.barplot(data=skill_df, x='Count', y='Skill', hue='Skill', palette='viridis', legend=False)
plt.title('Top 15 Trending Skills in 2024')
plt.show()
```

### 2. Job Seniority Distribution

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='job_level', hue='job_level', order=df['job_level'].value_counts().index, palette='magma', legend=False)
plt.title('Distribution of Job Seniority Levels')
plt.show()
```

### 3. Geographic Hubs

```python
top_locations = df['job_location'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_locations.values, y=top_locations.index, hue=top_locations.index, palette='rocket', legend=False)
plt.title('Top 10 Hiring Locations')
plt.show()
```
